```python
def llistaSelectivaOrd(llista, n):
    posicio = recbinsearch(n, llista, 0, len(llista) - 1)
    return llista[posicio:len(llista)]


def recbinsearch(x, nums, low, high):
    if low >= high:  # fem que retorni si només queda un
        if nums[low] > x:
            return low
        else:
            return low + 1  # no volem incloure el num trobat
    mid = (low + high) // 2
    items = nums[mid]
    if items == x:
        return mid + 1
    elif x < items:
        return recbinsearch(x, nums, low, mid-1)
    else:
        return recbinsearch(x, nums, mid+1, high)


llistaSelectivaOrd([10, 11, 12], 10)
```

```python
def factorial(num):
    factorial = 1
    if num < 0:
        print("Entra un enter positiu!")
    elif num == 0:
        print("0! = 1")
    else:
        for i in range (1, num+1):
            factorial *= i
    return factorial

factorial(8)
```

```python
from typing import NoReturn


def divPrimers(num):
    divPrim=[]

    for i in range (1,num+1):
        if num%i==0:
            primer = True
            for j in range (2,i-1):
                if i%j==0: 
                    primer = False
                    break
            if  primer: divPrim.append(i)
    return (divPrim)


def girar(num):
    num_string= str(num)
    n= len(num_string)
    rev= num_string[n::-1]
    num= int(rev)
    return (num)
    

def palindromics(nMax):
    cad=[]

    for i in range (1, nMax+1):
        gir= girar(i)

        div1= divPrimers(i)
        div2= divPrimers(gir)

        if div1==div2 and i!= gir: 
            cad.append(i)

    return(cad)

palindromics(5000)
```

```python
def camel(text):
    characters= "-_"

    #build array from text
    arr= list(text)

    # replace each char of the array after a -_ with its upper
    for i in range (len(arr)):
        if arr[i] in characters:
            arr[i+1]=arr[i+1].upper()
    

    # join array into text
    text2="".join(arr)

    # remove -_
    for x in range (len(characters)):
        text2=text2.replace(characters[x],"")

    return text2


print(camel("the-stealth_warrior"))
```

```python
def bbinary(n):
    # we build an array for the binary number
    nBin=[]

    # we build the binary number
    while n > 1 :
        res=n%2
        nBin.append(res)
        n //= 2
    nBin.append(n)
    
    nBin.reverse()

    # count how many 1 it has
    count=0
    for i in range (len(nBin)):
        if nBin[i]==1:
            count += 1
    

    return(count)

```

```python
def uniqueOrder(text):

    # we convert the text into an array and we build an empty array
    arr=list(text)
    cad=[]
    if len(arr)==0 : return (arr)

    # we check the elements and include them into the array
    j=0
    n=len(arr)
    while j<n-1:
        if arr[j] != arr[j+1]:
            cad.append(arr[j])
        j+=1

    # we check the last element
    if len(cad) == 0: return(arr)
    if arr[n-1]!= cad[len(cad)-1]:
        cad.append(arr[n-1])

    return(cad)

uniqueOrder("A")
```

```python
import math

def perfectSquare(sq):
    quad= int(math.sqrt(sq))
    seg = quad + 1

    if quad*quad == sq: 
        return (seg*seg)
    else:
        return (-1)

perfectSquare(1)
```
